```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv("sales_data.csv")
df=pd.DataFrame(data)
df
```

```python
reg_sale=df.groupby(['Region'])['Sales'].sum().reset_index()
reg_sale
```

```python
max_sales=reg_sale['Sales'].max()
reg_sale['Highlight']=['Max' if sale == max_sales else 'Normal' for sale in reg_sale['Sales']]


plt.figure(figsize=(10,6))
plt.tight_layout()
sns.barplot(data=reg_sale,x='Region',
    y='Sales',hue='Highlight',palette={'Normal':'lightgreen','Max':'darkgreen'}, edgecolor='black')
```

```python
plt.figure(figsize=(10,6))
plt.tight_layout()
sns.lineplot(data=reg_sale,x='Region',
    y='Sales',hue='Highlight',palette={'Normal':'lightgreen','Max':'darkgreen'})
```

```python

plt.figure(figsize=(10,6))
plt.tight_layout()
sns.scatterplot(data=reg_sale,x='Region',
    y='Sales',hue='Highlight',palette={'Normal':'lightgreen','Max':'darkgreen'}, edgecolor='black',markers='O')
```
